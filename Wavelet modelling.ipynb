```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tmin =-0.2
tmax = 0.2
Nmax =2000
fmin =5
fmax =25
Df =5
t = np.linspace(tmin,tmax,Nmax)    #time array
F = list(range(fmin,fmax+1,Df))    #list of frequencies
```

```python
for i in range(len(F)):
    r = (1-2*np.pi**2*F[i]**2*t**2)*np.exp(-np.pi**2*F[i]**2*t**2)
    plt.plot(t,r)
    plt.grid()
    plt.title("A ricker wavelet for frequency, f =%s Hz" %F[i])
    plt.xlim([-0.25,0.25])
    plt.xlabel("time")
    plt.ylabel("Amplitude")
    plt.show()
```
